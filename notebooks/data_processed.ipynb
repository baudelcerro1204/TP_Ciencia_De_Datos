```python
from pathlib import Path
import os

# ¿En qué carpeta estoy?
print("CWD:", os.getcwd())

# ¿Qué archivos hay en el raw?
raw = Path('../data/processed')  # sube un nivel para llegar al data/raw real
print("data/processed contiene:", [f.name for f in raw.iterdir()])
```

```python
import pandas as pd

# 1. Carga del Parquet procesado (ruta relativa CORRECTA)
df = pd.read_parquet('../data/processed/spotify_processed.parquet', engine='pyarrow')

# 2. Vista rápida
print("➤ Primeras 5 filas:")
display(df.head())

print("\n➤ Información de tipos y no nulos:")
df.info()

print("\n➤ % de valores faltantes por columna:")
missing = (df.isna().sum() / len(df)).sort_values(ascending=False)
display(missing.to_frame('pct_missing'))

print("\n➤ Estadísticas descriptivas de las métricas numéricas:")
display(df.describe().T)
```

```python
import pandas as pd
from pathlib import Path

# 1) Confirmar carpeta y fichero
print("CWD:", Path().resolve())
print("Enriched parquet:", Path('data/processed/spotify_enriched.parquet').exists())

# 2) Cargar el parquet enriquecido
df = pd.read_parquet('../data/processed/spotify_enriched.parquet', engine='pyarrow')

# 3) Primer vistazo
print("➤ Primeras 5 filas:")
display(df.head())

# 4) Info de tipos y nulos
print("\n➤ Información general:")
df.info()
print("\n➤ % de nulos por columna:")
missing = (df.isna().sum() / len(df) * 100).sort_values(ascending=False)
display(missing.to_frame('pct_missing'))

# 5) Estadísticas descriptivas
print("\n➤ Estadísticas numéricas:")
display(df.describe().T)

# 6) Análisis rápido de géneros y streams
print("\n➤ Recuento por género:")
display(df['genre'].value_counts())

print("\n➤ Top 10 álbumes por total_streams:")
display(df.sort_values('total_streams', ascending=False)
          [['artist_name','album_name','total_streams']].head(10))
```
