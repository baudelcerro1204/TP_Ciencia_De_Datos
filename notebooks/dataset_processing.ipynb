```python
# TP_Ciencia_de_Datos/notebooks/MillionSongDataset_AED.ipynb

# Celda 1: Imports y setup
import sys
from pathlib import Path
sys.path.append(str(Path('../src/songrecommender').resolve()))

import pandas as pd
from processors.data_processing import MillionSongDataProcessor  # Aseg√∫rate que la clase est√© en este archivo
```

```python
# Celda 2: Procesar datos
processor = MillionSongDataProcessor()
df = processor.process_all()

print(f"Dataset combinado con {len(df)} filas y {len(df.columns)} columnas")
```

```python

# Celda 3: Vista previa de datos
print(df.columns.tolist())
df.head(15)
```

```python
# Celda 4: Estad√≠sticas b√°sicas
print("Estad√≠sticas num√©ricas:")
display(df.describe())
```

```python
# Celda 5: Conteo de g√©neros (puede tener nulos)
print("Top 10 g√©neros m√°s frecuentes:")
genre_counts = df['genre'].fillna('unknown').value_counts().head(10)
display(genre_counts)
```

```python
# Celda 6: Distribuci√≥n playcount (uso)
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.histplot(df['playcount'], bins=50, log_scale=True)
plt.title("Distribuci√≥n de playcount (log escala)")
plt.xlabel("Playcount")
plt.ylabel("Cantidad de reproducciones")
plt.show()
```

```python


# Celda 7: Top 10 artistas por n√∫mero de escuchas
top_artists = df.groupby('artist')['playcount'].sum().sort_values(ascending=False).head(20)
print("Top 20 artistas por playcount:")
display(top_artists)
```
