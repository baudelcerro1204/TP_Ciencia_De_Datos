```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar dataset JSON (un registro JSON por línea)
df = pd.read_json('../data/raw/900k Definitive Spotify Dataset.json', lines=True)
print("Columnas cargadas:", df.columns.tolist())
```

```python
# Primeras 5 filas
df.head()
```

```python
# Tipos, nulos, memoria
df.info()
```

```python
# Estadísticas descriptivas
df.describe()
```

```python
# Conteo de valores faltantes
df.isnull().sum()
```

```python
# Histograma de features clave
for col in ['Danceability','Energy','Positiveness','Tempo','Popularity']:
    plt.figure()
    df[col].hist()
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
# Matriz de correlación entre features y target
corr = df[['Danceability','Energy','Positiveness','Tempo','Popularity']].corr()
print(corr)
```
