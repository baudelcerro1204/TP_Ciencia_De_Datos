# EDA - Spotify Imputed Dataset
Este notebook realiza análisis exploratorio sobre `spotify_imputed.csv`.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)

# Cargar datos
df = pd.read_csv("../data/processed/spotify_imputed.csv")
df.head()
```

```python
# Información general
df.info()
```

```python
# Descripción estadística
df.describe().T
```

```python
# Distribuciones de variables clave
num_cols = df.select_dtypes(include="number").columns.tolist()
for col in ['danceability','energy','valence','tempo','popularity']:
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribución de {col}")
    plt.show()
```

```python
# Correlación
corr = df.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Matriz de Correlación")
plt.show()
```
