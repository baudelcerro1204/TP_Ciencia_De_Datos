```python
## 1. Carga de librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style="whitegrid", palette="muted")
```

```python
## 2. Carga del dataset
df = pd.read_csv('../src/data/raw/hotel_bookings.csv')
print("Shape:", df.shape)
df.head()
```

```python
## 3. Información general del dataset
df.info()
df.describe()
```

```python

## 4. Verificar valores nulos
nulls = df.isnull().sum()
nulls[nulls > 0]
```

```python
## 5. Distribución de cancelaciones
sns.countplot(data=df, x='is_canceled')
plt.title('Reservas Canceladas vs No Canceladas')
plt.xticks([0, 1], ['No Cancelada', 'Cancelada'])
plt.show()
```

```python
## 6. Cancelaciones por tipo de hotel
sns.countplot(data=df, x='hotel', hue='is_canceled')
plt.title('Cancelaciones por tipo de hotel')
plt.show()
```

```python
## 7. Análisis de la anticipación (lead time)
sns.histplot(data=df, x='lead_time', hue='is_canceled', bins=50, kde=True)
plt.title('Lead Time por tipo de reserva')
plt.show()
```

```python
## 8. Cancelaciones por mes
# Primero creamos una columna de mes ordenado
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
df['arrival_date_month'] = pd.Categorical(df['arrival_date_month'], categories=month_order, ordered=True)
cancelaciones_mes = df.groupby('arrival_date_month')['is_canceled'].mean().reindex(month_order)

cancelaciones_mes.plot(kind='bar', figsize=(10, 5))
plt.title('Porcentaje de cancelaciones por mes')
plt.ylabel('Tasa de cancelación')
plt.xticks(rotation=45)
plt.show()
```

```python
## 9. Cancelaciones por canal de distribución
sns.countplot(data=df, y='distribution_channel', hue='is_canceled')
plt.title('Cancelaciones por canal de distribución')
plt.show()
```

```python
## 10. Correlaciones numéricas
plt.figure(figsize=(12, 8))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Matriz de correlación')
plt.show()
```
