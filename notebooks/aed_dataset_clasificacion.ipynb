# EXTRAER 5 CANCIONES POR CLASE DE POPULARIDAD

```python
import pandas as pd
```

```python

# ✅ 1. Cargar el dataset
ruta_dataset = "../data/processed/dataset_clasificacion.csv"  # ← Cambiar si es necesario
df = pd.read_csv(ruta_dataset)
```

```python

# ✅ 2. Verificar columnas
print("Columnas del dataset:", df.columns.tolist())
assert "popularity_class" in df.columns, "Falta la columna 'popularity_class'"
```

```python
# ✅ 3. Ver distribución por clase
print("\nCantidad de canciones por clase:")
print(df["popularity_class"].value_counts())
```

```python
# ✅ 4. Tomar 5 canciones por clase
muestras = df.groupby("popularity_class").apply(lambda x: x.sample(n=5, random_state=42)).reset_index(drop=True)
```

```python

# ✅ 5. Mostrar resultados
print("\n🎵 Muestras extraídas:")
display(muestras)
```
