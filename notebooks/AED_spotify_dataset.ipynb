# EDA - Spotify Dataset CSV
Este notebook realiza un análisis exploratorio del archivo `spotify_dataset.csv`.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

```python
# Cargar el CSV
df = pd.read_csv(os.path.join("..", "data", "raw", "spotify_dataset.csv"))
print("Shape:", df.shape)
df.head()
```

```python
# Listado de columnas
df.columns.tolist()
```

```python
# Información del DataFrame
df.info()

# Valores nulos por columna
df.isnull().sum().sort_values(ascending=False)
```

```python
# Descripción estadística (numérica y categórica)
df.describe(include='all').T
```

```python
# Seleccionar variables numéricas
num_cols = df.select_dtypes(include=['number']).columns.tolist()
# Variables específicas de interés
plot_cols = [c for c in ['tempo','loudness','energy','danceability','popularity'] if c in num_cols]

for col in plot_cols:
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f'Distribución de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.show()
```

```python
# Matriz de correlación
if len(num_cols) > 1:
    corr = df[num_cols].corr()
    sns.heatmap(corr, annot=True, cmap='coolwarm')
    plt.title('Matriz de correlación de variables numéricas')
    plt.show()
```

```python
# Distribución de emociones y géneros
for col in ['emotion', 'genre']:
    if col in df.columns:
        plt.figure()
        sns.countplot(y=col, data=df, order=df[col].value_counts().index[:10])
        plt.title(f'Top 10 categorías en {col}')
        plt.show()
```

```python
# Conteo de etiquetas contextuales (Good for ...)
tag_cols = [c for c in df.columns if c.lower().startswith('good for')]
if tag_cols:
    counts = df[tag_cols].sum().sort_values(ascending=False)
    counts.plot(kind='barh')
    plt.title('Frecuencia de etiquetas contextuales')
    plt.xlabel('Número de canciones')
    plt.show()
```

```python
# Análisis de campo 'Similar Songs'
if 'Similar Songs' in df.columns:
    lens = df['Similar Songs'].dropna().apply(lambda x: len(x) if isinstance(x, (list, str)) else 0)
    lens.value_counts().sort_index().plot(kind='bar')
    plt.title('Número de songs similares listadas por track')
    plt.xlabel('Cantidad de similares')
    plt.ylabel('Frecuencia')
    plt.show()
```

```python
# Exportar resumen de estadísticas
os.makedirs(os.path.join('..','data','processed'), exist_ok=True)
df.describe().to_csv('../data/processed/spotify_dataset_summary.csv')
print('Resumen estadístico guardado en data/processed/spotify_dataset_summary.csv')
```
