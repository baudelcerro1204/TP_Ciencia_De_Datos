```python
# TP_Ciencia_de_Datos/notebooks/01_AED_msd_spotify_lastfm.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Ruta base de datos
data_dir = "../data/raw"
```

```python
# Archivos CSV principales
music_info_path = os.path.join(data_dir, "Music Info.csv")
user_history_path = os.path.join(data_dir, "User Listening History.csv")

# Cargar datasets
df_music = pd.read_csv(music_info_path)
df_history = pd.read_csv(user_history_path)
```

```python
# Mostrar info básica
print("=== Music Info ===")
print(df_music.info())
print(df_music.head())

print("\n=== User Listening History ===")
print(df_history.info())
print(df_history.head())
```

```python
# Análisis básico: Estadísticas descriptivas
print("\nMusic Info - Estadísticas descriptivas:")
print(df_music.describe(include='all'))

print("\nUser Listening History - Estadísticas descriptivas:")
print(df_history.describe(include='all'))
```

```python
# Visualización: Géneros más frecuentes
plt.figure(figsize=(12,6))
top_genres = df_music['genre'].value_counts().nlargest(10)
sns.barplot(x=top_genres.index, y=top_genres.values)
plt.title('Top 10 géneros en Music Info')
plt.xticks(rotation=45)
plt.show()
```

```python

# Visualización: Número de escuchas por usuario (histograma)
plt.figure(figsize=(10,6))
sns.histplot(df_history['playcount'], bins=50, log_scale=(False, True))
plt.title('Distribución de número de reproducciones por usuario')
plt.xlabel('Número de reproducciones')
plt.ylabel('Cantidad de registros (escala logarítmica)')
plt.show()
```
